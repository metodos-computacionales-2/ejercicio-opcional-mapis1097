```python
import matplotlib.pyplot as plt
import numpy as np
N = 100
r = 2
primer = []
segundo = []
tercero = []
cuarto = []
quinto = []
x = np.linspace (0,1,N)

#Se hacen los for para recorrer las cuatro listas
for i in range(0,len(x)):
    if(x[i]<0.5):
        primer.append(r*x[i])
    elif(x[i]>=0.5):
        primer.append(r*(1-x[i]))
            
for i in range(int(len(primer)/2)):
    result=np.log(np.abs(primer[i+1]-primer[i-1]/2))
    cuarto.append(result)
        
for i in range(0,len(x)):
    result_2=(np.sin(np.pi*x[i]))
    segundo.append(result_2)
    
for i in range(int(len(segundo)/2)):
    result_3=np.log(segundo[i+1]-segundo[i-1])
    tercero.append(result_3)
        
#Mapa logistico
def Logistic_map(x,L):
    return L*x*(1-x)


for i in range(int(len(Logistic_map(x,4))/2)):
    logaritmo=np.log(np.abs(Logistic_map(x,4)[i+1]-Logistic_map(x,4)[i-1]/2))
    quinto.append(logaritmo)
    
# se grafica
   
plt.figure(figsize=(15,5))
plt.subplot(1,6,1)
plt.plot(x,primer)
 
    
plt.subplot(1,6,2)
plt.plot((x,segundo))

    
plt.figure(figsize=(15,5))
plt.subplot(1,6,3)
plt.plot(np.linspace(0,1,50),tercero)

    
plt.subplot(1,6,4)
plt.plot(np.linspace(0,1,50),cuarto)

    
plt.figure(figsize=(15,5))
plt.subplot(1,6,5)
plt.plot(x,Logistic_map(x,4))


plt.subplot(1,6,6)
plt.plot(np.linspace(0,1,50),quinto)


plt.show()
```
